### Khách mời hội nghị

```python
from PIL import Image, ImageDraw, ImageFont
import os
from IPython.display import Image as IPImage, display

def render_text_pillow(text, font_size=12, font_path="arial.ttf"):
    # Check if font file exists
    if not os.path.exists(font_path):
        raise OSError(f"Font file not found: {font_path}")

    font = ImageFont.truetype(font_path, font_size)

    text_width, text_height = font.getsize(text)
    image_width = text_width + 20 # Adding padding
    image_height = text_height + 20 # Adding padding

    # Create an image with white background
    image = Image.new('RGBA', (image_width, image_height), (0, 0, 0, 0))
    # Create an image with white background
    draw = ImageDraw.Draw(image)

    # Load the font

    position = (10, 10)

    # Draw the text on the image
    draw.text(position, text, font=font, fill="white")

    # Save the image to a file
    image_path = "/content/rendered_text.png"
    image.save(image_path)

    return image_path

text = "Chị Đỗ Thị Mai"
font_size = 90
font_path = "/content/SVN-GilroyBold.otf"

render_text_pillow(text, font_size, font_path)




def create_video_with_text_list(video_path, text_list, font_size, font_path, output_folder, text_position):
# Ensure the output folder exists
    os.makedirs(output_folder, exist_ok=True)

    for text in text_list:
        # Render the text to an image
        text_image_path = render_text_pillow(text, font_size, font_path)

        # Load the video clip
        video = VideoFileClip(video_path)

        # Load the text image
        text_image = ImageClip(text_image_path).set_duration(video.duration)

        # Set the position and duration of the text image using pixel coordinates
        text_image = text_image.set_position(text_position).set_duration(video.duration)

        # Create a final video with the text overlay
        final_video = CompositeVideoClip([video, text_image])

        # Define output path for the current text
        output_path = os.path.join(output_folder, f"{text}.mp4")

        # Write the final video to a file
        final_video.write_videofile(output_path, codec="libx264")
```

### Có chức danh

```python
from PIL import Image, ImageDraw, ImageFont
import os
from moviepy.editor import VideoFileClip, CompositeVideoClip, ImageClip

def render_text_pillow(text_lines, font_sizes, font_paths, line_spacing=25):
    # Check if font files exist
    for font_path in font_paths:
        if not os.path.exists(font_path):
            raise OSError(f"Font file not found: {font_path}")

    fonts = [ImageFont.truetype(font_path, size) for font_path, size in zip(font_paths, font_sizes)]

    # Calculate the width and height required for the image
    max_line_width = max(font.getsize(line)[0] for font, line in zip(fonts, text_lines))
    total_height = sum(font.getsize(line)[1] for font, line in zip(fonts, text_lines)) + line_spacing * (len(text_lines) - 1)

    image_width = max_line_width + 20 # Adding padding
    image_height = total_height + 20 # Adding padding

    # Create an image with transparent background
    image = Image.new('RGBA', (image_width, image_height), (0, 0, 0, 0))
    draw = ImageDraw.Draw(image)

    # Draw each line of text, centered horizontally
    y = 10
    for i, (font, line) in enumerate(zip(fonts, text_lines)):
        line_width, line_height = font.getsize(line)
        x = (image_width - line_width) // 2 # Center horizontally
        draw.text((x, y), line, font=font, fill="white")
        y += line_height + line_spacing # Move to the next line with custom spacing

    # Save the image to a file
    image_path = "/content/rendered_text.png"
    image.save(image_path)

    return image_path

def create_video_with_text_list(video_path, text_list, font_sizes, font_paths, output_folder, text_position, numbers):
    # Ensure the output folder exists
    os.makedirs(output_folder, exist_ok=True)

    for number, text in zip(numbers, text_list):
        # Split text into multiple lines if necessary
        text_lines = text.split('\n')

        # Render the text to an image
        text_image_path = render_text_pillow(text_lines, font_sizes, font_paths)

        # Load the video clip
        video = VideoFileClip(video_path)

        # Load the text image
        text_image = ImageClip(text_image_path).set_duration(video.duration)

        # Set the position and duration of the text image using pixel coordinates
        text_image = text_image.set_position(text_position).set_duration(video.duration)

        # Create a final video with the text overlay
        final_video = CompositeVideoClip([video, text_image])

        # Define output path for the current text
        sanitized_text = f"{number}. {text}".replace(' ', '_').replace('\n', '_')
        output_path = os.path.join(output_folder, f"{sanitized_text}.mp4")

        # Write the final video to a file
        final_video.write_videofile(output_path, codec="libx264")
```
