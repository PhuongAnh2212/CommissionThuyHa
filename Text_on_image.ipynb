```python
# Function to place text on an image and save with "number_text.jpg" format for each name
def place_text_on_image(image_path, output_folder, names, start_number=55, horizontal_position=300, font_size=40, text_color=(0, 0, 0), padding_top=0):
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    for index, name in enumerate(names, start=start_number):
        image = Image.open(image_path)
        image_width, image_height = image.size  # Get image dimensions

        # Initialize ImageDraw
        draw = ImageDraw.Draw(image)

        # Load a font
        font = ImageFont.load_default()  # Use the default font

        try:
            # Optionally, use a custom font if available
            font = ImageFont.truetype("/content/MOTHERLAND-SIGNATURE.TTF", font_size)  # Replace with the path to your custom font file
        except IOError:
            print("Custom font not found, using default font.")

        # Calculate text bounding box and size
        bbox = draw.textbbox((0, 0), name, font=font)  # Get the bounding box of the text
        text_width = bbox[2] - bbox[0]  # Width of the text
        text_height = bbox[3] - bbox[1]  # Height of the text

        # Adjust horizontal position based on text width
        if text_width > 700:
            horizontal_position = (image_width - text_width) // 2 + 80  # Adding padding

        # Calculate horizontal position to center the text horizontally
        else:
            horizontal_position = (image_width - text_width) // 2

        # Calculate vertical position and apply padding from the top
        vertical_position = (image_height - text_height) // 2 - padding_top

        # Add text to the image
        draw.text((horizontal_position, vertical_position), name, font=font, fill=text_color)

        # Create the output file name in "number_name.jpg" format
        output_file_name = f"{index}_{name.replace(' ', '')}.jpg"
        output_path = os.path.join(output_folder, output_file_name)  # Full path for saving the image

        # Save the edited image
        image.save(output_path)

        # Display the image using matplotlib
        plt.imshow(image)
        plt.axis('off')  # Hide axes
        plt.show()

        print(f"Saved: {output_path}")

# Example usage
image_path = '/content/VPNT.png'  # Replace with your image path
names = [
    "Sn",
    "Thm",
    "Lu Hng",
    "Mr.M",
    "Phyllis & Trang",
    "Dung",
    "Chu & L",
    "Chin",
    "Vit",
    "Hng"
]

output_folder = '/content/drive/MyDrive/CKimAnh'

place_text_on_image(image_path,output_folder, names, start_number=1, horizontal_position=300, font_size=70, text_color=(0, 0, 0), padding_top=580)  # Black text
```

```python
from PIL import ImageFont, ImageDraw, Image

def get_text_size(text, font_size):
    # Load the font (you can specify your font file or use the default)
    try:
        font = ImageFont.truetype("/content/MOTHERLAND-SIGNATURE.TTF", font_size)  # Replace with your font path
    except IOError:
        font = ImageFont.load_default()  # Fallback to default font if custom font is not found

    # Create an image to draw on (just to get the size)
    image = Image.new('RGB', (1, 1))
    draw = ImageDraw.Draw(image)

    # Calculate text size using textbbox method
    bbox = draw.textbbox((0, 0), text, font=font)  # Get the bounding box of the text
    text_width = bbox[2] - bbox[0]  # Width of the text
    text_height = bbox[3] - bbox[1]  # Height of the text

    return text_width, text_height

# Example usage
text = "Trng Th Thanh Hng"
font_size = 70
width, height = get_text_size(text, font_size)

print(f"Text Width: {width}px, Text Height: {height}px")
```
