```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
```

```python
import os
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')
```

```python
# Function to place text on an image and save with "number_text.jpg" format for each name
def place_text_on_image(image_path, output_folder, names, horizontal_position=300, font_size=40, text_color=(0, 0, 0), padding_top=0):  # Black text
    for index, name in enumerate(names, start=1):
        image = Image.open(image_path)
        image_width, image_height = image.size  # Get image dimensions

        # Initialize ImageDraw
        draw = ImageDraw.Draw(image)

        # Load a font
        font = ImageFont.load_default()  # Use the default font

        try:
            # Optionally, use a custom font if available
            font = ImageFont.truetype("/content/MOTHERLAND-SIGNATURE.TTF", font_size)  # Replace with the path to your custom font file
        except IOError:
            print("Custom font not found, using default font.")

        # Calculate text bounding box and size
        bbox = draw.textbbox((0, 0), name, font=font)  # Get the bounding box of the text
        text_width = bbox[2] - bbox[0]  # Width of the text
        text_height = bbox[3] - bbox[1]  # Height of the text

        # Calculate horizontal and vertical positions to center the text
        horizontal_position = (image_width - text_width) // 2  # Center horizontally
        vertical_position = (image_height - text_height) // 2 - padding_top  # Center vertically
        # Add text to the image (horizontal at 300, vertically centered)
        draw.text((horizontal_position, vertical_position), name, font=font, fill=text_color)

        # Create the output file name in "number_name.jpg" format
        output_file_name = f"{index}_{name.replace(' ', '')}.jpg"

        # Save the edited image
        image.save(output_file_name)

        # Display the image using matplotlib
        plt.imshow(image)
        plt.axis('off')  # Hide axes
        plt.show()

        print(f"Saved: {output_file_name}")

# Example usage
image_path = '/content/VPNT.png'  # Replace with your image path
names = [
    "Vn",
    "Thoan",
    "Tng",
    "Thin",
    "Khim",
    "Hng",
    "Hoa",
    "Loan",
    "Hng",
    "Bnh",
    "Du",
    "Trinh",
    "Ngc",
    "Sn",
    "Tho",
    "Xun",
    "Tuyn",
    "Trm",
    "Hnh",
    "Doanh",
    "Tho",
    "Linh",
    "Hin"
]  # List of names
output_folder = '/content/drive/MyDrive/generated_images'  # Path to the folder where you want to save images

place_text_on_image(image_path,output_folder, names, horizontal_position=300, font_size=100, text_color=(0, 0, 0), padding_top=580)  # Black text
```
